```python
# Python promgram to demonstrate stuck implementation using list
# stack adalah strutktur data yang menyimpan data seperti tumpukkan dimana elemen terakhir yang dimasukkan dia yang keluar pertama. LIFO
stack = [] # membuat list kosong, nanti menjadi list biasa di python

# Append() function to push element in the stuck
stack.append('a') # append digunakan untuk menambahkan elemen ke akhir list
stack.append('i') # 'i' artinya dimasukkan ke dalam stack
stack.append('u')
stack.append('e')
stack.append('o') # berada ditumpukkan paling atas

print('initial stack') # tumpukkan awal
print(stack)

# pop()funtion to pop element from stack in LIFO order
print("\nElements popped from stack:") # element dikeluarkan dari tumpukkan
print(stack.pop()) # menghapus elemen terakhir 'o'
print(stack.pop()) # menghapus elemen 'e
print(stack.pop())

print('\nStack after elements are popped:')
print(stack) # print sisa stack

# ucommenting print(stack.pop()) will cause an IndexError as the stack is now empty
```

```python
# Stack dengan collection.deque

# Python program to demonstrate stack implementation using collection.deque
from collections import deque # double ended queue : struktur data yang bisa menambah atau menghapus element dari depan maupun belakang.
# mengambil class (deque) dari collections
stack = deque()

# Append() function to push element in the stack
# jika ingin menambah dari depan maka bisa diguakan stack.appendleft
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Initial stack:')
print(stack)

# Pop function to pop element from stack in LIFO order
print("\nElements popped from stack:")
print(stack.pop())
# jika ingin menghapus dari depan maka gunakan print(stack.popleft())
print(stack.pop())
print(stack.pop())

print('\nStack after elements are popped')
print(stack)


# ucommenting print(stack.pop()) will cause an IndexError as the stack is now empty
```

```python
# Stack dengan queque.LIFOQueue

# yang masuk dulu keluar dulu
from queue import LifoQueue  # menggunakan LifoQueue class dari modul queue

# Initializing a stack
stack = LifoQueue(maxsize = 5) # maksimum stack hanya 5 elemen

# qsize() show the number of elements in the stack
print(stack.qsize()) # qsize digunakan untuk mengetahui jumlah elemen dalam stack, karena masih koosng makanya 0

# put()function to push element in the stack
stack.put('a')
stack.put('i')
stack.put('u')
stack.put('e')
stack.put('o')

print("Full: ", stack.full()) # True karena stack masih kosong
print("Size: ", stack.qsize()) # megecek ukuran stack

# get() function to pop element from stack in LIFO order
print('nElemens popped from the stack')
print(stack.get()) # get() digunakan untuk mengambil stack sama seperti pop()
print(stack.get()) # diambil dari palaing belakang yaitu o baru e
print(stack.get())

print("\nEmpty: ", stack.empty()) # empty = kosong makanya false karena ada isinya
```

```python
# Stack dengan sigle linked list

class Node: # class ini digunakan untuk membuat satu node dalam linked list
  def __init__(self, value): # konstruktor yang dijalankan saat objek Node dibuat
    self.value = value # data yang disimpan
    self.next = None # node berikutnya

class Stack: # struktur data stack dengan single linked list


# Initializing a stuck, use a dummy node, which us erasier form handling edge cases.
  def __init__(self):
    self.head = Node("head") # penenda awal.
    self.size = 0

# String representation of the stack
  def __str__(self): # isi stack model str
    cur = self.head.next # cur → current, node sedang diperiksa
    out = "" # variabel yang menyimpan hasil stack
    while cur:
      out += str(cur.value) + "->" # menambahkan nilai node ke str output
      cur = cur.next # pindah ke node berikutnya
    return out[:-2] # mengembalikan hasil string dan menghapus tanda -> terkhir

# Get the current size of the tack
  def getSize(self): # mengetahui jumlah elemn
    return self.size # mengembalikan nilai jumlah

# Check if the stack is empty
  def isEmpty(self): # apakah stack kosong?
   return self.size == 0

# Get the top item of the stack
  def peek(self)  : # peek untuk mengetahui elemen paling atas stack tanpa menghapusnya
  # Samitary check to see if we are peeking an empty stack
    if self.isEmpty():
      raise Exception("Peeking from an empty stack") # eror jika stack kosong
    return self.head.next.value # mengembalikan nilai node paling atas

# Push a value into the stack
  def push(self, value):
    node = Node(value)
    node.next = self.head.next # node baru menunjuk node sebelumnya berada di atas stack
    self.head.next = node
    self.size += 1 # menambah jumlah stack

# Remove a value into the stack and return
  def pop(self):
    if self.isEmpty():
      raise Exception("Popping from an empty stack")
    remove = self.head.next
    self.head.next = self.head.next.next
    self.size -= 1 # mengurnagi jml elemn stack
    return remove.value

# Driver Code
if __name__=="__main__": # memastikan kode dijalankan jika hanya file dijalankan langsung
    stack = Stack() # membuat stack baru
    for i in range(1, 11):
          stack.push(i)
    print(f"Stack: {stack}") # menampilkan isi stack

    for _ in range(1, 6):
        remove = stack.pop()
        print(f"Pop: {remove}")
    print(f"Stack: {stack}")
```
