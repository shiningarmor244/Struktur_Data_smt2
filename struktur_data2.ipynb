```python
# Tugas Praktikum 2

# LATIHAN 2
# 1. Buatlah double linked list dengan 3 node didalamnya, kemudian lakukan penyisipan tengah dan hapus dengan kata kunci !
# 2. Buatlah circular linked list dengan 3 node didalamnya, kemudian lakukan penyisipan akhir dan hapus akhir !


# Create Node

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None


# Buat 3 node
n1 = Node("Senin")
n2 = Node("Selasa")
n3 = Node("Rabu")

# Hubungkan manual (DOUBLE)
n1.nextval = n2
n2.prevval = n1

n2.nextval = n3
n3.prevval = n2

head = n1   # head menunjuk ke node pertama


# Fungsi print maju
def print_forward(head):
    temp = head
    while temp:
        print(temp.dataval)
        temp = temp.nextval


# Fungsi print mundur
def print_backward(head):
    temp = head
    while temp.nextval:
        temp = temp.nextval

    while temp:
        print(temp.dataval)
        temp = temp.prevval


print("Data Awal:")
print_forward(head)

print("\nData dari belakang:")
print_backward(head)

# Penyisipan tengah (setelah Selasa)
print("\nMenyisipkan'Kamis' setelah 'Selasa'")

newNode = Node("Kamis")

temp = head
while temp:
    if temp.dataval == "Selasa":
        break
    temp = temp.nextval

if temp:
    newNode.nextval = temp.nextval
    newNode.prevval = temp

    if temp.nextval:
        temp.nextval.prevval = newNode

    temp.nextval = newNode

print_forward(head)


# Hapus berdasarkan kata kunci ("Selasa")
print("\nMenghapus 'Selasa'")

temp = head
while temp:
    if temp.dataval == "Selasa":
        break
    temp = temp.nextval

if temp:
    if temp.prevval:
        temp.prevval.nextval = temp.nextval
    else:
        head = temp.nextval

    if temp.nextval:
        temp.nextval.prevval = temp.prevval

print_forward(head)


```

```python
# SOAL 2

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class CircularLinkedList:
    def __init__(self):
        self.headval = None

    # Tambah di akhir
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)

        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = NewNode
            return

        last = self.headval
        while last.nextval != self.headval:
            last = last.nextval

        last.nextval = NewNode
        NewNode.nextval = self.headval

    # Hapus di akhir
    def RemoveEnd(self):

        if self.headval is None:
            return

        # Jika hanya 1 node
        if self.headval.nextval == self.headval:
            self.headval = None
            return

        last = self.headval
        while last.nextval.nextval != self.headval:
            last = last.nextval

        last.nextval = self.headval

    # Print circular list
    def listprint(self):
        if self.headval is None:
            return

        temp = self.headval
        while True:
            print(temp.dataval)
            temp = temp.nextval
            if temp == self.headval:
                break



Li = CircularLinkedList()

Li.AddInEnd("Senin")
Li.AddInEnd("Selasa")
Li.AddInEnd("Rabu")

print("Data awal:")
Li.listprint()

Li.AddInEnd("Kamis")

print("\nSetelah penyisipan akhir:")
Li.listprint()

Li.RemoveEnd()

print("\nSetelah hapus akhir:")
Li.listprint()
```
