```python
# Menggunakan prinsip FIFO - first in first out

# Endqueue → menambahkan data ke antrian  → menggunakan append()

# Dequeue → menghapus data dari antrian → menggunakan pop()


# Queue dengan List

# Python program to demonstate queue implememntation using list
queue = [] # membuat list kosong bernama queue

# append() function to enqueue elemnt in the queue
queue.append('a') # menambahkan data ke queue
queue.append('i') # karena ini menggunakan list maka untuk push masih menggunakan append()
queue.append('u')
queue.append('e')
queue.append('o')

print("initial queue")
print(queue) # queue a → i → u → e → o

# pop() function to duque element from queue in FIFO order
print('\nElements dequeue from queue: ') # menghapus elemen
print(queue.pop(0)) # karena masih berbentuk list maka untuk menghapus masih menggunakan pop()
print(queue.pop(0)) # karena ini anntrian jadi yang masuk dulu yang keluar lebih dulu
print(queue.pop(0))

print('\nQueue after removing elemnts')
print(queue)

# uncommenting print(queue.pop()) will cause an IndexError as the queue is now empty

```

```python
#  Queue dengan collections.deque
# Python program to demontrate queue implementation using collection.deque

# deque → Doubly Ended Queue → antrian yang bisa menambah atau menghapus elemen di depan dan belakang dengan cepat.
from collections import deque # mengimport atau mengambil deque dari collections

queue = deque() # membuat queue kosong → queue adalah objek dari deque

# append() function to Enqueue elemnt in the queue
queue.append('a') # menambhakan elemen ke queue
queue.append('i') # kalau queue antrian selalu dimulai dari a (brisan depan)
queue.append('u')
queue.append('e')
queue.append('o') # 'o' berada dibarisan paling belakang

print('Initial queue: ')
print(queue)

# popleft() function to dequeue elemnt from stack in FIFO order
print("\nElements dequeued from the queue")
print(queue.popleft()) # menghapus elemen dari depan queue
print(queue.popleft()) # sesuai aturan yang keluar atau yang dihapus 'a' dulu baru 'i'
print(queue.popleft()) # kalau ingin meghapus dari belakang misal 'o' maka tetap menggunakan print(queue.pop())

print("\nQueue after removing elements")
print(queue)

# uncommening print(queue.popleft()) will cause an IndexError as the queue is now empty
```

```python
# Queue dengan queque.queue
# kelas bawaan dari modul queue yang yang digunakan untuk membuat queue yang benar-benar FIFO
# aman digunakan jika ada beberapa tread yang menambahkan atau menghapus secara bersamaan.

# Python program to demonstate queue implementation using queue module
from queue import Queue # mengimport queue dengan mengambil dari modul Queue

# Initializing a queue
qu = Queue(maxsize = 5) # membuat queue dengan maksimal 5 elemen

# qsize() show the number of elements in the queue
print(qu.qsize()) # menunjukkan jumlah elemen → karena baru dibuat dan belum ada isisnya maka output 0

# put() fuction to enqueue element in the queue
qu.put('a')  # put sama dengan append berfungsi untuk meambahkan elemen
qu.put('i')
qu.put('u')
qu.put('e')
qu.put('o') # 'o' berada di urutan paling belakang

print("Full: ", qu.full()) # akan bernilai True jika queue penuh
print("Size: ", qu.qsize()) # menghitung jumlah elemen sekarang setelah diisi → ooutput 5

# get() function to dequeue element from queue
print("\nElements dequeud from the queue")
print(qu.get()) # fungsinya sama dengan pop() yaitu untuk mengambil atau menghapus
print(qu.get()) # yang dihapus selalu dari urutan paling pertama masuk yaitu 'a'
print(qu.get())

print("\nEmpty: ", qu.empty()) # akan bernilai True jika queue kosong tapi akan bernilai Falsh jika ada isisnya
print("Full: ", qu.full()) # akan bernilai True jika queue full atau penuh
```

```python
# Queue dengan Single Linked List

class Queue: # membuat kelas Queue sendiri
  def __init__(self): # constructor, dipanggil saat objek Queue dibuat
    self.queue = list() # membuat list kosong untuk menampung elemen yang akan ditambahkan

  def addtoqu(self, dataval): # fungsi untuk menambah elemen → add (menambahkan) to (ke) qu (queue) → hanya nama fungsi (bebas)
# Insert method to add elment
    if dataval not in self.queue: # mencegah duplikasi
      self.queue.insert(0, dataval) # menambahkan elemen di index 0, artinya elemen baru masuk di depan bukan di belakang
      return True # menandakan apakah penambahan nanti berhasil atau tidak
    return False
# Pop method to remove elment
  def removefromq(self): # dequeue → menghapus elemen → fromq cuma nama
    if len(self.queue)>0: # menghitung berapa banyak elemen → .0 cek apakah queue tidak kosong
      return self.queue.pop() # kalau ada elemen hapus(pop)
    return ("No elements in Queue") # kalau kosong muncul pesan "No ...."

  def size(self): # mengmbalikan jumlah elemen di queue saat ini
    return len(self.queue)

TheQueue = Queue()
print(TheQueue.addtoqu("Jan")) # True, ditambahkan
# addtoqu seperti menaruh barang ke rak
print(TheQueue.addtoqu("Feb")) # True
# add (menambahkan)
print(TheQueue.addtoqu("March")) # True
print(TheQueue.addtoqu("April")) # True
print(TheQueue.size()) # ada 4 elemet → output = 4
print(TheQueue.removefromq()) # 'Jan' keluar (sesuai FIFO)
# removefromq seperti mengambil barang dari rak
print(TheQueue.removefromq()) # 'Feb' keluar
print(TheQueue.size()) # jumlah elemen setelah 'Jan' dan 'Feb keluar tinggal 2 tadinya 4
```
